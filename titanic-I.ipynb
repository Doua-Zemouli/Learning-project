```python
import pandas as pd
```

```python
# Get the data to Python environment and making a pandas data frame
df = pd.read_csv('/home/dell/Téléchargements/Data/jnotebook/train.csv')
df
```

```python
#filtering irrelevant columns
relevant_columns = ['Survived', 'Pclass', 'Age']
df2=df.filter(items = relevant_columns)
```

```python
# Display the first 10 rows of the new DataFrame
print(df2.head(10))
```

```python
# Display basic information about the DataFrame
print("Basic Information:")
print(df2.info())

# Display summary statistics
print("\nSummary Statistics:")
print(df2.describe())

# Display number of missing values in each column
print("\nNumber of Missing Values:")
print(df2.isnull().sum())
```

```python
# Calculate the mean of the Age column
age_mean = df2['Age'].mean()

# Replace missing values in the Age column with the mean
#df2['Age'].fillna(age_mean, inplace=True) this was the old way of doing it 
df2.fillna({'age': age_mean}, inplace=True)

# Verify that there are no missing values in the Age column
print("\nNumber of Missing Values in Age column:")
print(df2['Age'].isnull().sum())
```

```python
# Round the values in the Age column
df2['Age'] = df2['Age'].round()

# Display summary statistics
print("\nSummary Statistics:")
print(df2.describe())
```

```python
# Round the values in the Age column
df2['Age'] = df2['Age'].round()

# Define a function to categorize age
def categorize_age(age):
    if age <= 17:
        return 'Children'
    elif 18 <= age <= 64:
        return 'Adults'
    else:
        return 'Seniors'

# Create the 'age_category' column
df2['age_category'] = df2['Age'].apply(categorize_age)

print(df2['age_category'].value_counts())
```

```python
# Export the DataFrame to an Excel file
df2.to_excel('output2.xlsx', index=False)

print("\nData has been exported to output2.xlsx")
```
